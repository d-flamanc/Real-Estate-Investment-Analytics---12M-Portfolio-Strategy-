## 1) Show me the market trends to help me decide. Which US real-estate market is doing best overall. 10 points

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
from scipy.stats import linregress

# Construct the CSV export URL from the sheet URL
sheet_url = 'https://docs.google.com/spreadsheets/d/1GjAAX3lv3JpFPr3_eUTH_XI_nPNOO783Q_BJvcNjgTk/edit?usp=drive_link'
csv_export_url = sheet_url.replace('/edit?usp=drive_link', '/export?format=csv')

# Read the CSV data into a Pandas DataFrame
df = pd.read_csv(csv_export_url)

# Display the first few rows
display(df.head())

# Check for NaN values
print("Number of NaN values in each column:")
print(df.isna().sum())

# Display rows with NaN values
nan_rows = df[df.isna().any(axis=1)]
print("Rows with NaN values:")
display(nan_rows)

# Distinct states with NaN values
print("Distinct 'State' names with NaN values:")
print(nan_rows['State'].unique())

# Clean currency columns
df['HomeValue_Subgroup_Index_USD'] = df['HomeValue Subgroup Index USD'].replace('[\\$,]', '', regex=True).astype(float)
df['HomeValue'] = df['HomeValue'].replace('[\\$,]', '', regex=True).astype(float)

# Drop Iowa
df = df[df['State'] != 'IA']

# Histogram and Box Plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))
sns.histplot(df['HomeValue'], bins=30, kde=True, ax=ax1)
ax1.set_title('Histogram of HomeValues')
ax1.set_xlabel('Home Value')
ax1.set_ylabel('Frequency')
ax1.grid(True)

sns.boxplot(x=df['HomeValue'], ax=ax2)
ax2.set_title('Box and Whisker Plot of HomeValues')
ax2.set_xlabel('Home Value')
ax2.grid(True)

plt.tight_layout()
plt.savefig('histogram_and_boxplot_home_values.png')
plt.show()

# Home Value by Year and Subgroup
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='HomeValue_Subgroup_Index_USD', hue='Subgroup', errorbar=None)
plt.title('Home Value by Year and Subgroup')
plt.xlabel('Year')
plt.ylabel('Home Value Subgroup Index USD')
plt.grid(True)
plt.legend(title='Subgroup')
plt.savefig('home_value_by_year_and_subgroup.png')
plt.show()

# Outlier subgroup
outlier_df = df[df['Subgroup'] == 'Outlier']

# Mean Home Value by Year and Region
mean_region_values_by_year_outlier = outlier_df.groupby(['Year', 'REGION'])['HomeValue'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.lineplot(data=mean_region_values_by_year_outlier, x='Year', y='HomeValue', hue='REGION', errorbar=None)
plt.title('Mean Home Value by Year and Region (Outlier)')
plt.xlabel('Year')
plt.ylabel('Home Value')
plt.grid(True)
plt.legend(title='REGION')
plt.savefig('mean_home_value_by_year_and_region_outlier.png')
plt.show()

# West Region Outliers
west_region_outlier_df = outlier_df[outlier_df['REGION'] == 'West']

# Home Value by Year and State
plt.figure(figsize=(12, 8))
sns.lineplot(data=west_region_outlier_df, x='Year', y='HomeValue', hue='State', errorbar=None)
plt.title('Home Value by Year and State for REGION = West (Outlier)')
plt.xlabel('Year')
plt.ylabel('Home Value')
plt.grid(True)
plt.legend(title='State')
plt.savefig('home_value_by_year_and_state_west_outlier.png')
plt.show()

# YoY Growth %
west_region_outlier_df['YoY_Growth'] = west_region_outlier_df.groupby('State')['HomeValue'].pct_change() * 100
plt.figure(figsize=(12, 8))
sns.lineplot(data=west_region_outlier_df, x='Year', y='YoY_Growth', hue='State', errorbar=None)
plt.title('YoY Growth % by Year and State for REGION = West (Outlier)')
plt.xlabel('Year')
plt.ylabel('YoY Growth %')
plt.grid(True)
plt.legend(title='State')
plt.savefig('yoy_growth_by_year_and_state_west_outlier.png')
plt.show()

# Trend Analysis
significant_decrease_states_outlier = []
strong_growth_states_outlier = []

for state in west_region_outlier_df['State'].unique():
    state_data = west_region_outlier_df[west_region_outlier_df['State'] == state]
    slope, intercept, r_value, p_value, std_err = linregress(state_data['Year'], state_data['HomeValue'])
    if slope < 0 and p_value < 0.05:
        significant_decrease_states_outlier.append(state)
    if slope > 0 and r_value > 0.7:
        strong_growth_states_outlier.append(state)

print("States with statistically significant decrease in home value (Outlier):")
print(significant_decrease_states_outlier)

# Top 10 States by Avg YoY Growth
average_yoy_growth_outlier = west_region_outlier_df.groupby('State')['YoY_Growth'].mean().reset_index()
top_10_states_yoy_growth_outlier = average_yoy_growth_outlier.sort_values(by='YoY_Growth', ascending=False).head(10)

fig, ax1 = plt.subplots(figsize=(12, 8))
sns.barplot(x='State', y='YoY_Growth', data=top_10_states_yoy_growth_outlier, ax=ax1)
ax1.set_title("Top 10 States with Highest YoY HomeValue Growth % in West REGION (Outlier Subgroup)")
ax1.set_xlabel('State')
ax1.set_ylabel('YoY Growth %')

ax2 = ax1.twinx()
ax2.plot(top_10_states_yoy_growth_outlier['State'], top_10_states_yoy_growth_outlier['YoY_Growth'].cumsum(), color='red', marker='o')
ax2.set_ylabel('Cumulative YoY Growth %')

for i in range(len(top_10_states_yoy_growth_outlier)):
    ax2.text(i, top_10_states_yoy_growth_outlier['YoY_Growth'].cumsum().iloc[i], round(top_10_states_yoy_growth_outlier['YoY_Growth'].cumsum().iloc[i], 2), color='red')

fig.tight_layout()
plt.savefig('pareto_plot_top_10_states_yoy_growth_outlier.png')
plt.show()

# Wyoming Cities
wyoming_df = west_region_outlier_df[(west_region_outlier_df['State'] == 'WY') & (west_region_outlier_df['YoY_Growth'] > 0)]
average_yoy_growth_wyoming = wyoming_df.groupby('CITY')['YoY_Growth'].mean().reset_index().sort_values(by='YoY_Growth', ascending=False)
print("Cities driving positive YoY Growth % in Wyoming:")
print(average_yoy_growth_wyoming)

# Colorado Cities
colorado_df = west_region_outlier_df[(west_region_outlier_df['State'] == 'CO') & (west_region_outlier_df['YoY_Growth'] > 0)]
average_yoy_growth_colorado = colorado_df.groupby('CITY')['YoY_Growth'].mean().reset_index().sort_values(by='YoY_Growth', ascending=False)
print("Cities driving positive YoY Growth % in Colorado:")
print(average_yoy_growth_colorado)
````

#### General US Home Sales Trend ( Macro Level)

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load datasets
zhvi_df = pd.read_csv("https://drive.google.com/uc?id=1bAOIYxTuMknLU4WXqUiFT_mLPYhjjy-F&export=download")
listings_df = pd.read_csv("https://drive.google.com/uc?id=1PaNhHidgoXKtoLikWiZs3KhSH4m2ZME_&export=download")


# Filter for United States data
zhvi_us = zhvi_df[zhvi_df['RegionName'] == 'United States'].copy()
listings_us = listings_df[listings_df['RegionName'] == 'United States'].copy()

# Prepare time series
zhvi_ts = zhvi_us.iloc[0, 5:].T
zhvi_ts.index = pd.to_datetime(zhvi_ts.index)
zhvi_ts = zhvi_ts.rename("ZHVI")

listings_ts = listings_us.iloc[0, 5:].T
listings_ts.index = pd.to_datetime(listings_ts.index)
listings_ts = listings_ts.rename("New Listings")

# Calculate trendlines
def calculate_trendline(series):
    x = np.array(series.index.map(pd.Timestamp.toordinal)).reshape(-1, 1)
    y = series.values
    model = LinearRegression().fit(x, y)
    trend = model.predict(x)
    return trend

zhvi_trend = calculate_trendline(zhvi_ts)
listings_trend = calculate_trendline(listings_ts)

# Plotting
plt.figure(figsize=(14, 6))

# ZHVI plot
plt.subplot(1, 2, 1)
plt.plot(zhvi_ts.index, zhvi_ts.values, label='ZHVI', color='green')
plt.plot(zhvi_ts.index, zhvi_trend, label='Trendline', linestyle='--', color='black')
plt.title('U.S. Home Values Over Time')
plt.xlabel('Date')
plt.ylabel('ZHVI ($)')
plt.legend()
plt.grid(True)

# New Listings plot
plt.subplot(1, 2, 2)
plt.plot(listings_ts.index, listings_ts.values, label='New Listings', color='blue')
plt.plot(listings_ts.index, listings_trend, label='Trendline', linestyle='--', color='black')
plt.title('U.S. New Listings Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Listings')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

### Conclusion:
From a micro level, it can be shown that the prices all over all the regions of the United States were increasing with the West having a more steep increase.
From a macro level, it can be shown with positive slope trendline that the prices have been generally increasing with in the United States and this could be due to a fall in the number of new listing on the market which increases the demand hence a higher price.


## 3) Should I wait to invest or should I invest now and if so then where should I invest. (You could recommend more than one place)? 10 points

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Re-create merged_df to ensure it contains the 'HomeValue' column
# This assumes home_value_df and heat_index_df are already loaded and in scope from previous cells
home_value_df['RegionID'] = pd.to_numeric(home_value_df['RegionID'], errors='coerce').astype('Int64')
heat_index_df['RegionID'] = pd.to_numeric(heat_index_df['RegionID'], errors='coerce').astype('Int64')
merged_df = pd.merge(home_value_df, heat_index_df, on='RegionID', how='inner')

# --- Step 3: Clean HomeValue column ---
merged_df['HomeValue'] = pd.to_numeric(merged_df['HomeValue'].replace('[$,]', '', regex=True), errors='coerce')

# --- Step 4: Identify Outliers using IQR ---
Q1 = merged_df['HomeValue'].quantile(0.25)
Q3 = merged_df['HomeValue'].quantile(0.75)
IQR = Q3 - Q1
outlier_condition = (merged_df['HomeValue'] < Q1 - 1.5 * IQR) | (merged_df['HomeValue'] > Q3 + 1.5 * IQR)
outliers_df = merged_df[outlier_condition].copy()

# --- Step 5: Get Top 10 Outlier Markets by Avg. Home Value ---
market_scores = outliers_df.groupby('RegionName')['HomeValue'].mean().reset_index()
top10_outliers = market_scores.sort_values(by='HomeValue', ascending=False).head(10)
print(" Top 10 Outlier Markets by Average Home Value:\n")
print(top10_outliers)

# --- Step 6: Prepare Heat Index Columns ---
heat_cols = ['2020-12-31', '2021-12-31', '2022-12-31', '2023-12-31', '2024-12-31', '2025-02-28']
heat_index_clean = heat_index_df[['RegionID', 'RegionName', 'StateName'] + heat_cols].copy()

# Rename columns for readability
heat_index_clean = heat_index_clean.rename(columns={
    "RegionName": "City",
    "StateName": "State",
    "2020-12-31": "2020",
    "2021-12-31": "2021",
    "2022-12-31": "2022",
    "2023-12-31": "2023",
    "2024-12-31": "2024",
    "2025-02-28": "2025"
})

# --- Step 7: Merge Top 10 Outlier Cities with Heat Index ---
top10_df = pd.DataFrame({'City': top10_outliers['RegionName']})
top10_merged = pd.merge(top10_df, heat_index_clean, on='City', how='left')
top10_merged.set_index('City', inplace=True)

# --- Step 8: Plot Heatmap of Heat Index for Top Outliers ---
heatmap_data = top10_merged[['2020', '2021', '2022', '2023', '2024', '2025']].astype(float)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='coolwarm', linewidths=0.5, cbar_kws={'label': 'Heat Index'})
plt.title('Heat Map  Top 10 Outlier Markets by Home Value (20202025 Heat Index)')
plt.xlabel('Year')
plt.ylabel('City')
plt.tight_layout()
plt.show()
````

### Conclusion:

From the Macro level of US Home Value, it is best to invest now as Home Value eventually increases over time in the long run (refer to Q2 plots for General and Region US Home Value trends).

For a Micro level, the best places to invest in would be locations with above 70 on a heat map for their strong sellers market (CA: San Jose & San Francisco). Places with over 70 index on the Heat Map indicate that they have a "strong sellers market" as provided by Zillow (refer to Q1 plots).

We can also see that Santa Cruz had a strong sellers market in 2020 and 2021, but it dipped down to have a neutral market in 2025. **Importing the Rental Data From Zillow And Adding Average Monthly Rental Cost for each Year**

````python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)

# Updated URL to download the CSV directly
url = "https://drive.google.com/uc?export=download&id=1JV0-fk2_IjxLiAxtqwlR01X5GOj2Y1H1"

rental_df = pd.read_csv(url)

# Identify monthly rent columns
date_columns = rental_df.columns[5:]

#Convert columns names to datetime to extract years
date_years = pd.to_datetime(date_columns).year

# Map each year to its corresponding monthly column
year_to_columns = {}
for col, year in zip(date_columns, date_years):
    year_to_columns.setdefault(year, []).append(col)

#Compute average rent for each year and add it as a new column
for year, columns in year_to_columns.items():
    rental_df[f'AverageRent_{year}'] = rental_df[columns].mean(axis=1)

# Drop the original monthly rent columns
rental_df = rental_df.drop(columns = date_columns)

#Display head
rental_df.head()
````

**Plotting Average Rent change in the US over the last 10 years**

````python
# Define x as the years from 2015 to 2025
x = list(range(2015, 2026))

# Get the row for the United States
us_data = rental_df[rental_df["RegionName"] == "United States"]

# Extract the average rent values for those years
y = [us_data[f'AverageRent_{year}'].values[0] for year in x]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title("Average Rent in the US (20152025)")
plt.xlabel("Year")
plt.ylabel("Average Rent (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
````

**Find the House Value Average for each year and dropping columns that are not needed**

````python
import pandas as pd
import numpy as np


# Identify monthly home value columns (after metadata columns)
monthly_columns = shv_df.columns[5:]

# Convert column names to datetime and extract the year
date_years = pd.to_datetime(monthly_columns, errors='coerce').year

# Map each year to its corresponding monthly columns
year_to_columns = {}
for col, year in zip(monthly_columns, date_years):
    if pd.notnull(year):
        year_to_columns.setdefault(year, []).append(col)

# Calculate average home value for each year and add as new columns
for year, columns in year_to_columns.items():
    shv_df[f'AverageValue_{year}'] = shv_df[columns].mean(axis=1)

# Identify the yearly average columns
average_columns = [col for col in shv_df.columns if col.startswith('AverageValue_')]
years = sorted([int(col.split('_')[-1]) for col in average_columns])

# Define CAGR calculation function
def calculate_cagr(row):
    available = [(year, row[f'AverageValue_{year}']) for year in years if pd.notnull(row[f'AverageValue_{year}'])]
    if len(available) < 2:
        return np.nan
    start_year, start_value = available[0]
    end_year, end_value = available[-1]
    if start_value <= 0 or end_value <= 0 or end_year == start_year:
        return np.nan
    return (end_value / start_value) ** (1 / (end_year - start_year)) - 1

# Apply CAGR to each row
shv_df['AnnualValueGrowthRate'] = shv_df.apply(calculate_cagr, axis=1)

# Get rid of the US and Vineyard Haven
shv_df = shv_df[~shv_df['RegionName'].isin(['United States', 'Vineyard Haven, MA'])]

# Identify monthly columns: starting from column 6 and formatted like dates
monthly_columns = shv_df.columns[5:]
monthly_columns = [col for col in monthly_columns if pd.to_datetime(col, errors='coerce') is not pd.NaT]

# Drop the monthly value columns
shv_df = shv_df.drop(columns=monthly_columns)
shv_df
````

**Calculating the ROI after one year**

````python
import pandas as pd

# Rent projections (monthly rent values)
rent_df = pd.DataFrame({
    'RegionName': ['San Francisco, CA', 'San Jose, CA', 'Urban Honolulu, HI', 'Santa Cruz, CA', 'Key West, FL',
                   'Heber, UT', 'Edwards, CO', 'Jackson, WY', 'Breckenridge, CO'],
    'Rent_2026': [4007.87, 4387.89, 3418.15, 4305.40, 4335.33, 4969.96, 5275.92, 8545.87, 4350.62]
})

# Estimated 2025 home values
home_2025 = {
    'San Francisco, CA': 1245114, 'San Jose, CA': 1812202, 'Urban Honolulu, HI': 1057260,
    'Santa Cruz, CA': 1197260, 'Key West, FL': 1047523, 'Heber, UT': 1166868,
    'Edwards, CO': 1538678, 'Jackson, WY': 1573182, 'Breckenridge, CO': 1383272
}

# Assumptions
interest_rate = 0.06             # 6% mortgage interest (interest-only)
maintenance_rate = 0.015         # 1.5% one-time maintenance/tax cost
upgrade_one_time = 20000         # one-time upgrade cost

# ROI calculation for 1 year
roi_list_1yr = []

for i, row in rent_df.iterrows():
    region = row['RegionName']
    buy_price = home_2025[region]

    # Rent for 2026
    rent_2026 = row['Rent_2026'] * 12

    # One year of mortgage (interest-only)
    mortgage_1yr = (buy_price * interest_rate) / 12 * 12

    # One-time maintenance and upgrades
    maintenance_cost = buy_price * maintenance_rate
    upgrade_cost = upgrade_one_time

    # Net profit and ROI
    net_income = rent_2026 - mortgage_1yr - maintenance_cost - upgrade_cost
    roi = net_income / buy_price

    roi_list_1yr.append({
        'RegionName': region,
        'Rent_2026': round(rent_2026, 2),
        'Mortgage_1yr': round(mortgage_1yr, 2),
        'OneTimeMaint': round(maintenance_cost, 2),
        'OneTimeUpgrade': round(upgrade_cost, 2),
        'NetProfit': round(net_income, 2),
        'ROI_1yr (%)': round(roi * 100, 2)
    })

# Create DataFrame
roi_df_1yr = pd.DataFrame(roi_list_1yr).sort_values(by='ROI_1yr (%)', ascending=False)
roi_df_1yr
````
